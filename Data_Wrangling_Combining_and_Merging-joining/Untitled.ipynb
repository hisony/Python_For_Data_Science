```python

import pandas as pd  # pandas for data manipulation
import numpy as np   # numpy for numerical operations
```

```python
# Creating the first DataFrame, df1 with dictionary
#column name bole dewa lage na,dictionary er key gulai column name hoi
df1 = pd.DataFrame({'key': ['a', 'b', 'c', 'd', 'e'],
                    'A1': range(5),
                    'B1': range(5, 10)})

# Creating the second DataFrame, df2 with dictionary
df2 = pd.DataFrame({'key': ['a', 'b', 'c'],
                    'A2': range(3),
                    'B2': range(3, 6)})

```

```python
# Displaying df1 to see its contents
df1
 
```

```python
# Displaying df2 to see its contents
df2
```

```python
# Displaying information about df1
df1.info()
```

```python
# Performing an inner join on df1 and df2 using 'key' as the joining key
# Selecting only 'key', 'A1', and 'A2' columns from the merged result
#how diye ki dhoroner merge
#on diye common field
#ki ki dekhte chai pashe 3rd bracket diye 
pd.merge(df1, df2, how='inner', on='key')[['key', 'A1', 'A2']]
```

```python
# Displaying information about df1 again to see if there are any changes (there shouldnâ€™t be)
df1.info()
```

```python
# Checking the type of np.nan (used to represent missing values in pandas)
type(np.nan)
```

```python
# Performing a left join on df1 and df2 using 'key' as the joining key
# This will include all rows from df1(left table er sob gula row asbe), and matching rows from df2 where available
df4 = pd.merge(df1, df2, how='left', on='key')
```

```python
# Displaying df4 to see the result of the left join
#kono column e sob data type same thakbe.NULL er type by default float tai baki gulaw float
df4
```

```python
# Performing a right join on df1 and df2 using 'key' as the joining key
# This will include all rows from df2(right table theke sob row asbe), and matching rows from df1 where available
pd.merge(df1, df2, how='right', on='key')
```

```python
# Performing an outer join on df1 and df2 using 'key' as the joining key
# This will include all rows from both df1 and df2, with NaN where there are no matches
pd.merge(df1, df2, how='outer', on='key')
```

```python
# Creating the left DataFrame with multiple key columns (key1 and key2)
left = pd.DataFrame({'key1': ['a', 'a', 'b', 'c'],
                     'key2': ['a', 'b', 'a', 'b'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

    
# Creating the right DataFrame with multiple key columns (key1 and key2)
right = pd.DataFrame({'key1': ['a', 'b', 'b', 'c'],
                      'key2': ['b', 'a', 'a', 'a'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# Displaying the left DataFrame to see its contents
left
```

```python
# Displaying the right DataFrame to see its contents
right
```

```python
# Performing an inner join on left and right using both 'key1' and 'key2' as joining keys
# Only rows where both keys match in both DataFrames will be included
#joto gula jaigai key1 key2->a b,b a ase seigula show korbe 
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

```python
# Performing an outer join on left and right using both 'key1' and 'key2' as joining keys
# This will include all rows from both DataFrames, with NaN where there are no matches
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
# Performing a left join on left and right using both 'key1' and 'key2' as joining keys
# This will include all rows from left, and matching rows from right where available
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
# Performing a right join on left and right using both 'key1' and 'key2' as joining keys
# This will include all rows from right, and matching rows from left where available
pd.merge(left, right, how='right', on=['key1', 'key2'])
```
